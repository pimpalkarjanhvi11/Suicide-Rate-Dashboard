Cleaning of Suicide Rates Data in Python

```python
import pandas as pd
import numpy as np
df = pd.read_csv("/Users/janhvipimpalkar/Desktop/Cleaning/master.csv")
pd.set_option("display.max_columns", None)
df
```

Remove the E+ values from gdp_for_year($) and convert to 10^

```python

# Function to remove 'E+' part from numbers
def remove_num(x):
    if "E+" in x:
        # Split the string into parts before and after 'E+'
        parts = x.split("E+")
        
        # Extract the base and exponent parts
        base_part = parts[0]
        exponent_part = parts[1]
        
        # Convert the base part to a float and format it without exponential notation
        base_part_formatted = format(float(base_part), '.0f')
        
        # Concatenate the formatted base part and the exponent part
        result = base_part_formatted + " * 10^" + exponent_part
        
        return result
    else:
        return x

# Check and remove leading/trailing whitespaces in column names
df.columns = df.columns.str.strip()

# Apply the updated function to the column
df['gdp_for_year ($)'] = df['gdp_for_year ($)'].apply(remove_num)

# Print the updated DataFrame
df['gdp_for_year ($)']
```

Round each float value in the column to the nearest integer and convert it to an integer data type.

```python
import numpy as np
# Remove or replace non-finite values
df['suicides_no'] = df['suicides_no'].replace([np.inf, -np.inf], np.nan).fillna(0).astype(int)

# Print the cleaned column
df['suicides_no']
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
